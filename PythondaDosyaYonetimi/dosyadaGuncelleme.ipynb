"r+"  hem okuma hemde yazma yapmak anlamına gelir.

```python
with open("newfile.txt","r+", encoding="utf-8") as file:
  file.write("deneme")
with open("newfile.txt","r+", encoding="utf-8") as file:
  print(file.read())
```

```python
with open("newfile.txt","r+", encoding="utf-8") as file:
  file.seek(20)
  file.write("deneme")
with open("newfile.txt","r+", encoding="utf-8") as file:
  print(file.read())
```

sayfa sonunda güncelleme

```python
with open("newfile.txt","a", encoding="utf-8") as file:
  file.write("\nEmel Turan")
with open("newfile.txt","r", encoding="utf-8") as file:
  print(file.read())
```

sayfa başında güncelleme

```python
with open("newfile.txt","r+", encoding="utf-8") as file:
  content = file.read()
  content = "busray\n" + content
  file.seek(0)
  file.write(content)
with open("newfile.txt","r", encoding="utf-8") as file:
  print(file.read())
```

sayfa ortasında güncelleme

```python
with open("newfile.txt","r+", encoding="utf-8") as file:
  list = file.readlines()
  help(list.insert)
  list.insert(1,"ecrin yıldırım\n")
  file.seek(0)
  for i in list:
    file.write(i)
with open("newfile.txt","r", encoding="utf-8") as file:
  print(file.read())
```

```python
with open("newfile.txt","r+", encoding="utf-8") as file:
  list = file.readlines()
  help(list.insert)
  list.insert(1,"fatih yıldırım\n")
  file.seek(0)
  file.writelines(list)
with open("newfile.txt","r", encoding="utf-8") as file:
  print(file.read())
```

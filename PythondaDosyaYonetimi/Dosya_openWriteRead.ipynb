dosya açmak ve oluşturmak için open() fonksiyonu kullanılır. Kullanımı: open(dosya_adi, dosya_erişme_modu)

dosya_erisme_modu => dosyayı hangi amaçla açtığımızı belirtir.

"w": (Write) yazma modu. dosyayı konumda oluşturur.

"a": (Append) ekleme. dosya konumda yoksa oluşturur.

"x": (Create) oluşturma. Dosya zaten varsa hata verir.

"r": (Read) okuma. varsayılan dosya konumda yoksa hata verir.

```python
try:
  file = open("newfile.txt","r")
  #print(file)
  print("dosya okundu")
except FileNotFoundError:
  print("dosya okuma hatası")
finally:
  print("dosya kapandı.")
  file.close()
```

```python
file = open("newfile.txt","r",encoding="utf-8")
for i in file:
  print(i, end="")
```

```python
with open("newfile.txt","r", encoding="utf-8") as file:
  content = file.read()
  print(content)
```

```python
with open("newfile.txt","r", encoding="utf-8") as file:
  content = file.read()
  print(content)
  print(file.tell())  #tell() fonksiyonu, dosya işlemleri sırasında bir dosya imlecinin (file pointer) bulunduğu konumu (byte cinsinden) belirten bir metoddur.
  content2 = file.read()
  print(content2)
```

```python
with open("newfile.txt","r", encoding="utf-8") as file:
  content = file.read()
  print(content)
  file.seek(10)   #seek() fonksiyonu, bir dosya içindeki dosya imlecini (file pointer) belirli bir konuma taşımak için kullanılan bir metodudur.
  print(file.tell())
  content2 = file.read()
  print(content2)
```

```python
with open("newfile.txt","r", encoding="utf-8") as file:
  content = file.read(23)
  print(content)
  file.seek(0)
  print(file.tell())
  content2 = file.read()
  print(content2)
```

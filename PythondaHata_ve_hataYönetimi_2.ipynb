# 2: Kullanıcı 'q' değerini girmedikçe aldığınız her inputun sayı olduğundan emin olunuz aksi halde hata mesajı yazın.

```python
while True:
  sayi = input('sayı: ')
  if sayi == 'q':
     break
  try:
    result = float(sayi)
    print('girdiğiniz ssayı: ',result)
    break
  except ValueError:
    print('gecersiz sayi')
    continue
```

# 3: Girilen parola içinde türkçe karakter hatası veriniz.

```python
def checkPassword(parola):
  turkce_karakterler = 'şçğüöıi'

  for i in parola:
    if i in turkce_karakterler:
      raise TypeError('Parola türkçe karalter iiçeremez.')
    else:
      pass
  print('geçerli parola')
```

```python
parola = input('parola: ')

try:
  checkPassword(parola)
except TypeError as err:
  print(err)
```

# 4: faktöriyel fonksiyonu oluşturup fonksiyona gelen değer için hata mesajları verin.

```python
def faktoriyel(x):
  x = int(x)

  if x < 0:
    raise ValueError('Negatif değer')
  result = 1

  for i in range(1, x+1):
    result *=  i

  return result
```

```python
for x in [5, 10, 20, -3, '10a']:
  try:
    y = faktoriyel(x)
  except ValueError as err:
    print(err)
    continue
  print(y)
```

```python
result = re.findall("[abc]", str) #girilen karakterleri string içinde arar ve buldularını ekrana yazar
result
```

```python
len(result)
```

```python
result = re.findall("[a-c]", str) # a dan c ye kadar tüm karakterleri str içinde arar ve ekrana yazar
result
```

```python
len(result)
```

```python
result = re.findall("[^a-s]", str) # a dan c ye kadar tüm karakterler hariç diğer karakterleri str içinde arar ve ekrana yazar
result
```

```python
result = re.findall("^a", str) # str değişkeni a ile başlıyor mu
result
```

```python
result = re.findall("^B", str) # str değişkeni B ile başlıyor mu
result
```

```python
result = re.findall("a$", str) # str değişkeni a ile bitiyor mu
result
```

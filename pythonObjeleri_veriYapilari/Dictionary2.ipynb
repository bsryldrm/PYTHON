```python
ogrenciler = {
    '120' : {
        'ad' : 'Ali',
        'soyad': 'Yılmaz',
        'telefon': '532 000 00 00'
    },

    '125' : {
        'ad' : 'Can',
        'soyad' : 'Korkmaz'
        'telefon': '532 000 00 01'
    },

    '128' : {
        'ad' : 'Volkan',
        'soyad' : 'Yükselen',
        'telefon': '532 000 00 02'
    },
}
```

```python
#bilgileri verilen öğrencileri kullanıcıdan aldığınız bilgilerle dictionary içinde saklayınız.

ogrenciler = {}
number = input("öğrenci no:")
name = input("öğrenci adı:")
surname = input("öğrenci soyad:")
phone = input("öğrenci telefon:")
```

```python
ogrenciler[number] = {
    "ad" : name,
    "soyad" : surname,
    "telefon" : phone
}
```

```python
ogrenciler
```

```python
#update ile yapımı
#update kullanarak aynı anda birden fazla veri eklenir
ogrenciler.update({
    number : {
        'ad' : name,
        'soyad' : surname,
        'telefon' : phone
    }
})
```

```python
ogrenciler
```

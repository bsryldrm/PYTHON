```python
#girilen sayıdan hangisi büyüktür

a = int(input("a sayı:"))
b = int(input("b sayı:"))

result = (a > b)
print(f'a: {a} b: {b} den büyüktür: {result}')
```

```python
#kullanıcıdan 2 vize (%60) ve final (%40) notunu alıp ortalama hesaplayınız.
#eğer ortalama 50 ve üstündeyse geçti değilse kaldı

vize1 = float(input("1.vize: "))
vize2 = float(input("2.vize: "))
final = float(input("final: "))

ortalama = (((vize1 + vize2) /2 ) * 0.6) + (final * 0.4)

print(f'not ortalamanız: {ortalama} ve dersten geçme durumunuz {ortalama >= 50.0}')
```

```python
#girilen bir sayının tek mi çift mi olduğunu yazınn
sayi = int(input('sayı: '))

tekcift = (sayi % 2 == 0)

print(f"girilen sayının çift olma durumu {tekcift}")
```

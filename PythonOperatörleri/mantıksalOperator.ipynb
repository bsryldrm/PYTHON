```python
#girilen bir sayının  0-100 arasında olup olmadığını kontrol ediniz
sayi = float(input("sayi giriniz:"))
result = (sayi > 0) and (sayi < 100)
result
```

```python
#girilen bir sayının pozitif çift sayı olup olmadığını ko
sayi = float(input("sayi giriniz:"))
result = (sayi > 0) and (sayi % 2 == 0)
result
```

```python
#email ve parola bilgileri ile giriş kontrolü yapınız
email = "email@python.com"
paralo = "tr1071"

girilenEmail = input("email: ")
girilenParola = input("parola: ")

result = (girilenEmail == email) and (girilenParola == paralo)
result
```

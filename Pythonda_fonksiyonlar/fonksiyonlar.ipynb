```python
def factorial(number):

  def inner_factorial(number):
    if number <= 1:
      return 1

    return number * inner_factorial(number -1)

  return inner_factorial(number)

print(factorial(4))
```

```python
def factorial(number):
  if not isinstance(number,int): #isinstance() fonksiyonu, bir nesnenin belirli bir sınıfa veya veri türüne ait olup olmadığını kontrol etmek için kullanılan bir Python fonksiyonudur.
    raise TypeError("tam sayı değer girmelisiniz")

  def inner_factorial(number):
    if number <= 1:
      return 1

    return number * inner_factorial(number -1)

  return inner_factorial(number)
try:
  print(factorial(-2.5))
except Exception as ex:
  print(ex)
```

pythonda fonksiyonlara parametre olarak gönder

```python
def toplama(a,b):
  return a+b
def cikarma(a,b):
  return a-b
def carpma(a,b):
  return a*b
def bolme(a,b):
  return a/b

def islem(f1,f2, f3, f4, islem_adi):
  if islem_adi == "toplama":
    print(f1(2,3))
  elif islem_adi == "cikarma":
    print(f2(5,3))
  elif islem_adi == "carpma":
    print(f3(3,4))
  elif islem_adi == "bolme":
    print(f4(10,2))
  else:
    print("geçersiz işlem...")
```

```python
islem(toplama, cikarma, carpma, bolme, "toplama")
islem(toplama, cikarma, carpma, bolme, "cikarma")
islem(toplama, cikarma, carpma, bolme, "carpma")
islem(toplama, cikarma, carpma, bolme, "bolme")
islem(toplama, cikarma, carpma, bolme, "bolmeee")
```

pythonda decorator fonksiyonları



---Python'da bir fonksiyonu değiştirmek veya genişletmek için kullanılan özel işlevlere "decorator" denir. Decorator'lar, fonksiyonları sarmalayarak (wrap) veya değiştirerek ek özellikler eklemek için kullanılır. Bu, kodunuzu daha temiz, daha modüler ve daha okunabilir hale getirmenize yardımcı olabilir.



```python
def my_decorator(func):
  def wrapper():
    print("fonksiyondan önceki işlemler")
    func()
    print("fonksiyondan sonraki işlemler")
  return wrapper
```

```python
@my_decorator
def sayHello():
  print("hello")
```

```python
sayHello()
```

```python
#ornek
def my_decorator(func):
  def wrapper(name):
    print("fonksiyondan önceki işlemler")
    func(name)
    print("fonksiyondan sonraki işlemler")
  return wrapper
```

```python
@my_decorator
def sayHello(name):
  print("hello",name)
```

```python
sayHello("büşra")
```

```python
#ornek
import math
import  time

def calculate_time(func):
  def inner(*args, **kwargs):
      start = time.time()
      time.sleep(1)

      func(*args, **kwargs)

      finish = time.time()
      print("fonksiyon "+ func.__name__ + " " + str(finish - start)+ "saniye sürdü.")
  return inner

@calculate_time
def usAlma(a,b):
  print(math.pow(a,b))

@calculate_time
def faktoriyel(num):
  print(math.factorial(num))
```

```python
usAlma(2,8)
```

```python
faktoriyel(4)
```
